```python
import csv
import pandas as pd
import numpy as np
R = 5  # 1 -- _d123; 2 -- _d137b; 3 -- _dlab; 4 -- _d205_wdw; 5 -- _d210_wdw


file = open("old_data/{}.csv".format(R))
read = csv.reader(file, delimiter=';')
_read = pd.DataFrame(list(read)).to_numpy()
data_start = 1
data_end = len(_read)
q = _read[:, 1][data_start:data_end].astype(np.float)
ti = _read[:, 2][data_start:data_end].astype(np.float)
te = _read[:, 3][data_start:data_end].astype(np.float)
dt = ti- te
file.close()
#q = q.reshape(-1, 1)
#ti = ti.reshape(-1, 1)
#te = te.reshape(-1, 1)
```

```python
import matplotlib.pyplot as plt
plt.plot(q)
plt.plot(dt)
```

```python
heading = ['v1', 'v2']

with open('old_data/d210_wdw.csv', 'w') as file:
    write = csv.writer(file)
    write.writerow(heading)
    write.writerows([dt[i], q[i]] for i in range(len(q)))
```
